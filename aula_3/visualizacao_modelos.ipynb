```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from random import *
```

```python
categories = np.genfromtxt("categories_places365.txt",delimiter=" ",dtype=None,encoding="utf-8")
categories = [name[0] for name in categories]
categories = categories[0:360:3]
categories[0]
```

```python
scores = pd.read_csv("model_scores.csv")
```

```python
def highlight(ax,highlight_indexes,highlight_color='#cc3333'):
    for bar in ax.patches:
        bar.set_facecolor('#d3d3d3')
    for idx in highlight_indexes:
        ax.patches[idx].set_facecolor(highlight_color)
```

```python
def find_indexes(df1,df2,start_idx=0,end_idx=1):
    highlight_indexes = []
    if end_idx==1:
        end_idx = df1.shape[0]
    for i in range(start_idx,end_idx):
        if df1[i] >= df2[i]:
            highlight_indexes.append(i)

    return highlight_indexes
```

```python
width = 0.35
idx = find_indexes(scores['model_a'],scores['model_b'])

plt.figure(figsize=(22,6))
ax1 = plt.bar(np.arange(len(categories)),scores['model_a'],width=width,label='model_A',color='#0333FF')
ax2 = plt.bar(np.arange(len(categories))+width,scores['model_b'],width=width,label='model_B',color='#03DAC5')
plt.xticks(np.arange(len(categories)), categories,fontsize=7, rotation=90)
plt.legend()

highlight(ax1,idx,'#0333FF')
highlight(ax2,idx,'#03DAC5')

plt.axhline(y=np.mean(scores['model_a']),linewidth=1, color='#FF0266',linestyle='dashed')

plt.show()
```
