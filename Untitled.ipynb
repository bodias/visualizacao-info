```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from random import *
from scipy.stats import bernoulli


```

```python
categories = np.genfromtxt("categories_places365.txt",delimiter=" ",dtype=None,encoding="utf-8")
categories = [name[0] for name in categories]
categories = categories[0:360:3]
categories[0]
```

```python
def sigmoid(x):
    return (1 / (1 + np.exp(-x)))

def highlight(ax,highlight_indexes):
    for bar in ax.patches:
        bar.set_facecolor('#d3d3d3')
    for idx in highlight_indexes:
        ax.patches[idx].set_facecolor('#cc3333')
```

```python
scores1 = np.random.exponential(size=120)
scores2 = bernoulli.rvs(size=120,p=0.7)
scores3 = np.random.normal(loc = 0.7, scale=np.sqrt(1) , size= 120) 
```

```python
scores1 = sigmoid(scores1)
scores3 = sigmoid(scores3)        
        
```

```python
scoreb_1 = np.random.beta(0.5,0.7,size=120)
print(scoreb_1.mean())
print(scoreb_1.min())
print(scoreb_1.max())

plt.hist(scoreb_1)
```

```python
print(scores1.mean())
print(scores1.min())
print(scores1.max())
plt.hist(scores1)
```

```python
#scores2 = sigmoid(scores2)
print(scores2.mean())
print(scores2.min())
print(scores2.max())
plt.hist(scores2)
```

```python
print(scores3.mean())
print(scores3.min())
print(scores3.max())
plt.hist(scores3)
```

```python
highlight_indexes = []
for i in range(0,120):
    if scores1[i] >= scores2[i]:
        highlight_indexes.append(i)
        
len(highlight_indexes)        
```

```python
plt.figure(figsize=(22,6))
ax = plt.bar(np.arange(len(categories)),scores1,width=0.85)
plt.xticks(np.arange(len(categories)), categories,fontsize=7, rotation=75)
highlight(ax,highlight_indexes)


plt.show()
```

```python
highlight_indexes = []
for i in range(0,120):
    if scores2[i] >= scores1[i]:
        highlight_indexes.append(i)
        
len(highlight_indexes)     
```

```python
plt.figure(figsize=(22,6))
ax = plt.bar(np.arange(len(categories)),scores2,width=0.9)
plt.xticks(np.arange(len(categories)), categories,fontsize=7, rotation=75)
highlight(ax,highlight_indexes)
plt.show()
```

```python
for idx in highlight_indexes:
    # ganha por pouca diferena
    if (scores2[idx] - scores2[idx]) < 0.1:
         scores2[idx] = 0.95 - scores2[idx]
```

```python
plt.figure(figsize=(22,6))
plt.bar(np.arange(len(categories)),scores3,width=0.9)
plt.xticks(np.arange(len(categories)), categories,fontsize=7, rotation=75)

plt.show()
```

```python
plt.boxplot([scores1,scores2,scores3,scoreb_1])
plt.ylim(ymin=0,ymax=1)
plt.show()
```
